Import the data the drop the 'casual' and 'registered' columns. These are redundant with your target, 'count'.

```python
#Load data
import pandas as pd
filename = "/content/drive/MyDrive/Week2_data/Machine Learning Data Sets/bikeshare_train - bikeshare_train.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df.drop(columns=['casual', 'registered'], inplace=True)
```

Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:

    Name of the Month
    Name of the Day of the Week
    Hour of the Day
        Make sure all 3 new columns are 'object' datatype so they can be one-hot encoded later.
        Drop the 'datetime' and 'season' columns. These are now redundant.


```python
df.dtypes
```

```python
# Convert the 'datetime' column to datetime type
df['datetime'] = pd.to_datetime(df['datetime'])
df.dtypes
```

```python
# Create new columns for month, year, and hour of the day
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year
df['hour_of_day'] = df['datetime'].dt.hour

# Print the data frame head
print(df.head())
```

```python
# Convert all three new columns to object datatype
df['month'] = df['month'].astype('object')
df['year'] = df['year'].astype('object')
df['hour_of_day'] = df['hour_of_day'].astype('object')
```

```python
# Drop the 'datetime' and 'season' columns
df = df.drop(['datetime', 'season'], axis=1)
```

```python
# Convert the 'temp' and 'atemp' columns to Fahrenheit using lambda and apply
df['temp'] = df['temp'].apply(lambda x: (x * 1.8) + 32)
df['atemp'] = df['atemp'].apply(lambda x: (x * 1.8) + 32)
```

```python
# Calculate the temperature variance
df['temp_variance'] = df['temp'] - df['atemp']
```
